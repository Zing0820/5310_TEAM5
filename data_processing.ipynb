### Raw data
- This is a large raw dataset consists of much information of one gracery store in New York.
-  From the column name like order, days_hour_of _day, product_name, department and so one, we can figure out people's potnetial habits when they do grocery shopping. 
-  These information will be a great help to research on opening a new grocery store in Broklyn. 

```python
data=pd.read_csv('ECommerce_consumer behaviour.csv')
data
```

### First of all, I deleted all data including "missing" and "other " since they will not provide meaning ful information.

```python
data=data.dropna()
mask = ~data.isin(['missing', 'other']).any(axis=1)
data= data[mask]
```

### Table 0 department  and its products
- For this table, I will not enter this into our database system. This table provide useful indormation when we do visualization afterwards.

```python
department= data[['department_id','department']]
```

```python
# Assuming 'department' and 'product_name' are the relevant columns
# Ensure both are string type
data['department'] = data['department'].astype(str)
data['product_name'] = data['product_name'].astype(str)

# Group by 'department' and join the 'product_name's
department_products = data.groupby('department')['product_name'].apply(lambda x: ', '.join(set(x))).reset_index()
# department_products.to_excel('department_product.xlsx',index=False)
department_products
```

### Table 1 department
-  I got department name  and department id in this table. There is no information regarding department_id 2.

```python
department= data[['department_id','department']]
department=department.drop_duplicates()
department=department.sort_values('department_id')
department.to_csv('department.csv',index=False)
department
```

### Table 2 product_dept
- We have product_id，product_name，department_id in this table.

```python
product_dept=data[['product_id','product_name','department_id']]
product_dept=product_dept.drop_duplicates()
product_dept=product_dept.sort_values('product_id')
product_dept.to_csv('product_dept.csv',index=False)
product_dept
```

### Table 3 product_order
- product_id，add_to_cart_order，order_id

```python
data.sort_values(by=['user_id', 'order_id'], inplace=True)
data['order_count'] = data.groupby('user_id').cumcount() + 1
data['cart_id'] = d=data['user_id'].astype(str) + '_' + data['order_count'].astype(str)
data.drop('order_count', axis=1, inplace=True)
```

```python
data
```

```python
product_order=data[['cart_id','product_id','add_to_cart_order']]
product_order=product_order.sort_values(['cart_id','add_to_cart_order'])
product_order
product_order.to_csv('product_order.csv',index=False)
# product_order[['cart_id']=='2_1']
```

### Table 4 user
- We have user_id and order_number here. 

```python
data
```

```python
import pandas as pd
unique_order_counts = data.groupby('user_id')['order_id'].nunique().dropna()
unique_order_counts = unique_order_counts.reset_index()
data1 = pd.merge(data, unique_order_counts, on='user_id', how='left')
data1.rename(columns = {'order_id_y':'order_number1', 'order_id_x':'order_id'}, inplace = True)
data1
order_number=data1[['user_id','order_number1']]
order_number=order_number.drop_duplicates()
order_number
order_number.to_csv('user.csv', index=False)
order_number
```

### Table 5 order
- We have order_id, user_id, cart_id, order_hour_of_day , days_since_prior_order.

```python
order=data[['order_id','user_id','cart_id','order_hour_of_day','days_since_prior_order']]
# 假设你的数据存在DataFrame df中
order = order.drop_duplicates()
order=order.sort_values('user_id')
order= order[order.columns[[1,0] + list(range(2, order.shape[1]))]]
order=order.drop_duplicates(subset=['user_id', 'order_id'])
order

cols = list(order.columns)
cols[0], cols[1] = cols[1], cols[0]
order=order[cols]
order
```

### Table 6 recorded
- In this table, we have user_id, order_id, product_id and reordered columns. 

```python
reorder=data[['user_id','order_id','product_id','reordered']]
reorder.drop_duplicates(subset=['user_id','order_id','product_id','reordered'])
reorder=reorder.drop_duplicates()
reorder =reorder.reset_index(drop=True)
reorder['Number'] = reorder.index + 1
reorder
cols = ['Number'] + [col for col in reorder if col != 'Number']
reorder = reorder[cols]
reorder
reorder.to_csv('reorder.csv',index=False)
reorder

# reorder[reorder['order_id']==2873174]
```
